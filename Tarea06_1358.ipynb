```
class NodoCircular:
  def __init__(self,dato):
    self.dato = dato
    self.siguiente = None

#Constructor Lista circular
class CircularList:
  def __init__(self):
    self.primero = None
    self.ultimo = None
    self.size = 0
  def get_size(self):
    return self.size

#Saber si esta vacia 
  def is_empty(self):
    if self.primero == None:
      print("La lista esta vacia")
    else:
      print("La lista no esta vacia")

#Insertar un nodo de forma ordenada
  def insert(self,dato):
    if self.primero == None and self.ultimo == None:
      n = NodoCircular(dato)
      self.primero = self.ultimo = n   
      self.ultimo.siguiente = self.primero
         
    else:
      n = NodoCircular(dato)
      if n.dato < self.primero.dato:
        recorrer = self.primero
        self.primero = n
        self.primero.siguiente = recorrer
        
      elif n.dato > self.ultimo.dato:
        r = self.ultimo 
        self.ultimo = n
        r.siguiente = n
        self.ultimo.siguiente = self.primero
        
      else:
        curr_node = self.primero
        while curr_node != None:                 
          if n.dato < curr_node.siguiente.dato:
            curr_node = curr_node
            r = curr_node.siguiente
            curr_node.siguiente = n
            n.siguiente = r            
            
            break            
          curr_node = curr_node.siguiente
    self.size += 1
          
          
 #Recorrido transversal         
  def transversal(self):
    curr_node = self.primero
    while curr_node.siguiente != None:
      print(curr_node.dato,"-->", end="")
      curr_node = curr_node.siguiente
      if curr_node == self.primero:
        break
    print("")

#Buscar un elemento a partir de su valor
  def search(self,dato):
    curr_node = self.primero
    while curr_node != None:
      if curr_node.dato == dato:
        print(curr_node.dato == dato)
        break
      curr_node = curr_node.siguiente
      if curr_node == self.primero:
        print(curr_node.dato == dato)
        break

#Eliminar un elemento a partir de un valor  
  def remove(self,dato):
    if self.primero.dato == dato:   
      nsp = self.primero.siguiente
      nsps = self.primero.siguiente.siguiente
      self.primero = nsp
      self.primero.siguiente = nsps
      self.ultimo.siguiente = self.primero

    elif self.ultimo.dato == dato:
      curr_node = self.primero
      while curr_node != None:
        if curr_node.siguiente.dato == dato:
          nsu = curr_node       
          self.ultimo = nsu
          self.ultimo.siguiente = self.primero
        curr_node = curr_node.siguiente
        if curr_node == self.primero:
          break

    else:
      curr_node = self.primero
      while curr_node != None:
        if curr_node.siguiente.dato == dato:
          ncs = curr_node.siguiente.siguiente
          curr_node.siguiente = ncs
        curr_node = curr_node.siguiente

        if curr_node == self.primero:
          break

```

```
cl = CircularList()

"""Â¿Esta vacia?"""
cl.is_empty()
print("")


"""Insertar"""
cl.insert(30)
cl.transversal()
print(cl.get_size())
cl.insert(40)
cl.transversal()
cl.insert(50)
cl.insert(20)
cl.insert(60)
cl.transversal()
cl.insert(25)
cl.transversal()
print(cl.get_size())
cl.insert(80)
cl.transversal()
print(cl.get_size())
print("")

"""Buscar"""
cl.search(25)
cl.search(85)
print("")

"""Remover"""
cl.transversal()
cl.remove(20)
cl.transversal()
cl.remove(80)
cl.transversal()
cl.remove(50)
cl.transversal()

```
